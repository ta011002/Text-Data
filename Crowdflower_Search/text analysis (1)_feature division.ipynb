# Text preprocessing

```python
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize # 문장 tokenize
from nltk.tokenize import word_tokenize # 단어 tokenize
from nltk.stem import LancasterStemmer # 어간 추출
from nltk.stem import WordNetLemmatizer # 원형 복원
from nltk.tag import pos_tag # POS tagging
```

## Tokenize

```python
%%time
for num in range(len(df_train["query"])):
    df_train["query"][num] = word_tokenize(df_train["query"][num])
```

```python
%%time
for num in range(len(df_train["product_title"])):
    df_train["product_title"][num] = word_tokenize(df_train["product_title"][num])
```

```python
%%time
for num in range(len(df_train["product_description"])):
    df_train["product_description"][num] = word_tokenize(df_train["product_description"][num])
```

```python
df_train
```

## Convert all text to lowercase

```python
%%time
for num in range(len(df_train['query'])):
    for i in range(len(df_train['query'][num])):
        df_train['query'][num][i] = df_train['query'][num][i].lower()
```

```python
%%time
for num in range(len(df_train['product_title'])):
    for i in range(len(df_train['product_title'][num])):
        df_train['product_title'][num][i] = df_train['product_title'][num][i].lower()
```

```python
%%time
for num in range(len(df_train['product_description'])):
    for i in range(len(df_train['product_description'][num])):
        df_train['product_description'][num][i] = df_train['product_description'][num][i].lower()
```
